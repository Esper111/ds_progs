```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

# Create a dictionary with applicant data (credit score and loan purpose)
data = {
    'CreditScore': [675, 690, 655, 660, 680, 675, 695, 685, 670, 690],
    'LoanPurpose': ['home improvement', 'debt consolidation', 'car purchase', 'medical', 'business',
                    'home improvement', 'debt consolidation', 'car purchase', 'medical', 'business'],
    'LoanApproval': ['yes', 'no', 'yes', 'yes', 'no', 'yes', 'no', 'yes', 'yes', 'no']
}

# Create a DataFrame from the dictionary
df = pd.DataFrame(data)

# Split features and labels
X = df[['CreditScore', 'LoanPurpose']]
y = df['LoanApproval']

# Convert loan purpose to numerical values using one-hot encoding
X_encoded = pd.get_dummies(X, columns=['LoanPurpose'], prefix='', prefix_sep='')

# Create and train the decision tree classifier
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_encoded, y)
```

```python
prediction_data = {
    'CreditScore': [675],
    'business': [False],
    'car purchase': [False],
    'debt consolidation': [False],
    'home improvement': [True],
    'medical': [False]
}

# Create a DataFrame with the prediction data
new_data = pd.DataFrame(prediction_data)

# Make predictions
prediction = clf.predict(new_data)

# Add prediction to the DataFrame
new_data['LoanApproval'] = prediction

# Display prediction
print(new_data)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Visualize the decision tree
plt.figure(figsize=(10, 6))
plot_tree(clf, feature_names=['CreditScore','business','car purchase','debt consolidation','home improvement','medical'], class_names=clf.classes_, filled=True)
plt.show()
```
