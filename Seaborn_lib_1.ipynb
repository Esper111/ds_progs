4.Survey analysis. What are the distribution and trends of responses to a satisfaction survey across different regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.DataFrame({'Region':['Bangalore','Chennai','Mumbai','Kolkata'],
                   'Response':[1,2,3,2]})
```

```python
sns.barplot(x='Region',y='Response',data=data,palette='muted')
plt.title('Responses based on region')
plt.xlabel('Region')
plt.ylabel('Response')
plt.show()
```
