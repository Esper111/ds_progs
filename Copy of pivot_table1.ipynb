Q1) In a sales analysis scenario, given a dataset containing sales transactions with columns for Date, Product Category, Quantity Sold, and Revenue, how would you use pandas pivot tables to identify the total quantity sold for each product category over different time periods, such as monthly or quarterly intervals?

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Generate sample data
np.random.seed(0)
random.seed(0)

# Generate dates for the past year
dates = pd.date_range(end=datetime.today(), periods=365, freq='D')

# Generate random product categories
categories = ['Electronics', 'Clothing', 'Books']
product_categories = [random.choice(categories) for _ in range(len(dates))]

# Generate random quantity sold and revenue
quantity_sold = np.random.randint(1, 100, size=len(dates))
revenue = np.random.randint(10, 500, size=len(dates))

# Create DataFrame
data = {
    'Date': dates,
    'Product Category': product_categories,
    'Quantity Sold': quantity_sold,
    'Revenue': revenue
}
df = pd.DataFrame(data)

# Save DataFrame to CSV
df.to_csv('sales_data.csv', index=False)
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv('sales_data.csv')


```

```python
# Convert 'Date' column to datetime type
df['Date'] = pd.to_datetime(df['Date'])
```

```python

# Set 'Date' column as the index
df.set_index('Date', inplace=True)

```

```python
# Define the frequency for grouping (e.g., 'M' for monthly, 'Q' for quarterly)
frequency = 'M'
```

```python

# Create pivot table to get total quantity sold for each product category over different time periods
pivot_table = pd.pivot_table(df, index=df.index.to_period(frequency), columns='Product Category',
                             values='Quantity Sold', aggfunc='sum', fill_value=0)
```

```python

# Display the pivot table
print(pivot_table)
```
