```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

# Create a dictionary with credit scores and loan approval
data = {
    'CreditScore': [750, 680, 800, 720, 690, 710, 650,705,683,657],
    'LoanApproval': ['yes', 'no', 'yes', 'yes', 'yes', 'yes', 'no','no','yes','no']
}

# Create a DataFrame from the dictionary
df = pd.DataFrame(data)

# Split features and labels
X = df[['CreditScore']]  # Features
y = df['LoanApproval']    # Labels

# Create and train the decision tree classifier
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X, y)

# Now, let's make a prediction for a new customer with a credit score of 700
prediction = clf.predict([[700]])

# Print the prediction
print("For a customer with a credit score of 700:")
print("Loan approval prediction:", prediction[0])
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Create and train the decision tree classifier
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X, y)

# Visualize the decision tree
plt.figure(figsize=(10, 6))
plot_tree(clf, feature_names=['CreditScore'], class_names=clf.classes_, filled=True)
plt.show()
```

```python
import graphviz
from sklearn.tree import export_graphviz

# Export the decision tree as a DOT file
dot_data = export_graphviz(clf, out_file=None,
                           feature_names=['CreditScore'],
                           class_names=clf.classes_,
                           filled=True)

# Visualize the decision tree
graph = graphviz.Source(dot_data)
graph.render("decision_tree")  # Save the decision tree to a file
graph.view()  # Open a viewer to display the decision tree
```
