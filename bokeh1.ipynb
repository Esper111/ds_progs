```python
from bokeh.plotting import figure, output_file, show
from bokeh.models import HoverTool, ColumnDataSource
import pandas as pd
```

```python
data={
    'x':[1,2,3,4,5],
    'y':[6,7,2,4,7],
    'size':[10,20,30,40,50],
    'color': ['navy', 'green', 'blue', 'red', 'orange'],
    'label': ['A', 'B', 'C', 'D', 'E']
}
df=pd.DataFrame(data)
#Create a ColumnDataSource from the DataFrame
source = ColumnDataSource(df)
```

```python
p = figure(title="Simple Scatter Plot", x_axis_label='X-Axis', y_axis_label='Y-Axis', tools="pan, wheel_zoom, box_zoom, reset, save")
#Add a scatter renderer with size, color, and alpha
scatter = p.scatter('x', 'y', size='size', color='color', alpha=0.5, source=source)
# Add hover tool
hover= HoverTool()
hover.tooltips = [
("Index", "$index"),
("(x, y)", "(@x, @y)"),
("Size", "@size"),
("Color", "$color[hex, swatch]:color"),
("Label", "@label")
]
p.add_tools(hover)
#Specify the output file
output_file("scatter_plot.html")
#Show the results
show(p)
```
