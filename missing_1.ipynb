1.You are provided with a  dataset containing customer data, but it has missing values in several columns. Using pandas fun how will u handle this missing vals to ensure accurate analysis and insights.


```python
import pandas as pd
import numpy as np
```

```python
data = {
    'CustomerID': [1, 2, 3, 4, 5],
    'Age': [25, np.nan, 35, np.nan, 45],
    'Gender': ['Male', 'Female', np.nan, 'Female', 'Male'],
    'Income': [50000, 60000, np.nan, 80000, 90000],
    'SpendingScore': [65, 80, 50, np.nan, 90]
}
df=pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
df.notna()
```

```python
df_new=df.dropna()
df_new
```

```python
df_new1=df.dropna(subset=["Gender"])
df_new1
```

```python
# df_n2=df_new1.interpolate(method='linear', inplace=True)
# df_n2
```

```python
df_new1.fillna(0)
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Income'] = df['Income'].fillna(df['Income'].mean())
df['SpendingScore'] = df['SpendingScore'].fillna(df['SpendingScore'].mean())
```
