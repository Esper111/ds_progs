You are tasked with building an image classification system to diff b/w diff types of animals in images. U hav a dataset of labeled images. How will u build SVM model to solve multiclass classification problem. Calculate accuracy, display  confusion matrix and classification report.

```python
import tensorflow_datasets as tfds
import tensorflow as tf
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler

# Load the TFDS "stanford_dogs" dataset
dataset, info = tfds.load('stanford_dogs', split='train', with_info=True)
```

```python
# Define function to preprocess images
def preprocess_img(sample):
    # Resize images to a fixed size (e.g., 100x100 pixels)
    image = tf.image.resize(sample['image'], [100, 100])
    # Normalize pixel values to range [0, 1]
    image = tf.cast(image, tf.float32) / 255.0
    label = sample['label']
    return image, label

# Apply preprocessing function to dataset
dataset = dataset.map(preprocess_img)
```

```python
# Extract images and labels
images = []
labels = []
for image, label in dataset:
    images.append(image.numpy())  # Convert to NumPy array
    labels.append(label.numpy())  # Convert to NumPy array

images = np.array(images)
labels = np.array(labels)

# Flatten images into feature vectors
X = images.reshape(images.shape[0], -1)

# Split dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=42)

# Feature scaling
# scaler = StandardScaler()
# X_train_scaled = scaler.fit_transform(X_train)
# X_test_scaled = scaler.transform(X_test)

# Train SVM model
svm_model = SVC()
svm_model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = svm_model.predict(X_test)
```

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Display confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)

# Display classification report
class_report = classification_report(y_test, y_pred)
print("Classification Report:")
print(class_report)
```
