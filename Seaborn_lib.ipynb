1.How do sales vary across different product catagories?


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.DataFrame({'Product Catagory':['Eletronics','Clothing','Books','Furniture'],
                   'Sales':[5000,7000,3000,6000]})
```

```python
sns.barplot(x='Product Catagory',y='Sales',data=data,palette='muted')
plt.title('Sales for different categories')
plt.xlabel('Product Catagory')
plt.xlabel('Sales')
plt.show()
```

2.Customer segmentation. Can we isulaize the distribution of customer ages within different income brackets?

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.DataFrame({'Income Bracket':['Low','Low','Medium','Medium','High','High'],
                   'Age':[25,30,35,40,45,50]})
```

```python
sns.violinplot(x='Income Bracket',y='Age',data=data, palette='muted')
plt.title('Customer with diff age for diff income bracket')
plt.xlabel('Income Bracket')
plt.ylabel('Age')
plt.show()
```

3.Market trends. How the stock market performance varied over different quarters of the year?

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.DataFrame({'Quarters':['Q1','Q2','Q3','Q4'],
                   'Stock performance':[1000,1050,1020,1150]})
```

```python
sns.lineplot(x='Quarters',y='Stock performance',data=data,marker='o')
plt.title('Stock market analysis')
plt.xlabel('Quarters')
plt.xlabel('Stock performance')
plt.show()
```
