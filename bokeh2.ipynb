1.How can we build a dynamic scatter plot that updates the displayed data based on user selected inputs from dropdown menus and sliders.

```python
import numpy as np

import matplotlib.pyplot as plt

from ipywidgets import interact, widgets
```

```python
np.random.seed(0)

x = np.random.rand(100)

y = np.random.rand(100)

colors = np.random.rand(100)

sizes = 1000 * np.random.rand(100)
```

```python
def scatter_plot(x_data, y_data, color, size, colormap='viridis', marker='o'):

  plt.figure(figsize=(10, 6))

  scatter = plt.scatter(x_data, y_data, c=color, s=size, cmap=colormap, marker=marker)

  plt.colorbar (scatter, label='Color scale')

  plt.xlabel('X axis')

  plt.ylabel('Y axis')

  plt.title('Dynamic Scatter Plot')

  plt.grid(True)

  plt.show()
```

```python
colormap_options = ['viridis', 'plasma', 'inferno', 'magma', 'cividis']
marker_options = ['o', 's', 'D', '^', 'v']

colormap_widget = widgets. Dropdown(

options=colormap_options,

value='viridis',

description='Colormap:'

)

marker_widget =widgets. Dropdown ( options=marker_options,

value='o',

description='Marker:'

)

alpha_widget =widgets. FloatSlider(

value=0.5,

min=0,

max=1.0,

step=0.1,

description= 'Alpha:'

)
```

```python
@interact(colormap=colormap_widget, marker=marker_widget, alpha=alpha_widget)

def update_plot(colormap, marker, alpha):

  scatter_plot(x, y, colors, sizes, colormap=colormap, marker=marker)
```
