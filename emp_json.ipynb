```python
import pandas as pd
```

```python
def load_emp_data():
  try:
    with open('/content/drive/MyDrive/datasets/emp_data.json','r') as file:
      data=json.load(file)
    df=pd.DataFrame(data)
    return df
  except FileNotFoundError:
    print("not found, creating one")
    return pd.DataFrame(columns=['ID','Name','Dept','Salary'])

def save_emp_data(df):
  data=df.to_dict(orient='records')
  with open('/content/drive/MyDrive/datasets/emp_data.json','w') as file:
    json.dump(data,file,indent=4)

def add_emp(df,new_emp):
  new_df=pd.DataFrame([new_emp])
  df=pd.concat([df,new_df],ignore_index=True)
  save_emp_data(df)
  return df

def update_emp(df,emp_id , field , new_value):
    df.loc[df['ID']==emp_id,field] = new_value
    save_emp_data(df)
    return df

def display_roster(df):
  print("Employee roster")
  print(df)

def main():
    employee_df = load_emp_data()
    display_roster(employee_df)
    new_employee = {"ID":101,'Name':'Emily','ept':'HR','Salary':60000}
    employee_df=add_emp(employee_df,new_employee)
    display_roster(employee_df)
    employee_df=update_emp(employee_df,101,'Salary',65000)
    display_roster(employee_df)

if __name__ =="__main__":
  main()
```
