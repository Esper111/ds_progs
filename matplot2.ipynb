```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Sample data generation
np.random.seed(42)  # For reproducibility
num_samples = 100

# Random x, y, z coordinates
x = np.random.uniform(-100, 100, num_samples)
y = np.random.uniform(-100, 100, num_samples)
z = np.random.uniform(-50, 50, num_samples)

# Simulated mineral concentration
concentration = np.random.uniform(0, 100, num_samples)
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Create scatter plot
scatter = ax.scatter(x, y, z, c=concentration, cmap='viridis')

# Add color bar
colorbar = fig.colorbar(scatter, ax=ax, pad=0.1)
colorbar.set_label('Mineral Concentration')

# Labels and title
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_zlabel('Depth')
ax.set_title('3D Spatial Distribution of Mineral Concentration')

# Show plot
plt.show()
```
