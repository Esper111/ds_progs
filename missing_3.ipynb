3. You are provided with a dataset containing user comments and posts from the platform, but the text data might contain inconsistencies such as missed spellings, punctuation variation and inconsistent capitalization. Using pandas string function, how do u preprocess the text data to standardize and extract relevant info fr analysis.

```python
import pandas as pd
import numpy as np
```

```python
data = {
    'User': ['User1', 'User2', 'User3', 'User4', 'User5'],
    'Comment': [
        'This is a great platform to learn new things. #excited',
        'I am having trouble with the Login process :(',
        'The customer support team was very helpful!',
        'i love the features Of this platform.',
        'Looking forward to the upcoming webinar!'
    ],
    'Post': [
        'Check out my Latest blog post about machine learning.',
        'New updates on the website are coming soon.',
        'excited to share my project on data visualization.',
        'Here are some tips for effective data analysis.',
        'Join us for an interactive session on AI and ML.'
    ]
}

df = pd.DataFrame(data)
df
```

```python
df['Comment'] = df['Comment'].str.lower()
df['Post'] = df['Post'].str.lower()
```

```python
df
```

```python
df['Comment'] = df['Comment'].apply(lambda x: ' '.join(word.capitalize() for word in x.split('. ')))
df['Post'] = df['Post'].apply(lambda x: ' '.join(word.capitalize() for word in x.split('. ')))
```

```python
df
```

```python
# cleaned_string = ''.join(char for char in sample_string if char not in string.punctuation)
import string
import re
def preprocess_text(text):
    # Convert text to lowercase
    text = text.lower()
    # Remove punctuation marks using string.punctuation
    text = ''.join(char for char in text if char not in string.punctuation)
    # Remove extra whitespaces
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# Apply preprocessing to 'Comment' and 'Post' columns
df['Comment'] = df['Comment'].apply(preprocess_text)
df['Post'] = df['Post'].apply(preprocess_text)
```

```python
df
```

```python
df['Comment Tokens'] = df['Comment'].str.split()
df['Post Tokens'] = df['Post'].str.split()
```

```python
df
```

```python
import nltk
from nltk.corpus import stopwords

# Download NLTK stopwords
nltk.download('stopwords')

# Load stopwords
stop_words = set(stopwords.words('english'))

df['Comment Tokens'] = df['Comment Tokens'].apply(lambda x: [word for word in x if word not in stop_words])
df['Post Tokens'] = df['Post Tokens'].apply(lambda x: [word for word in x if word not in stop_words])
```
