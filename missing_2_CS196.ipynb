2. You are provided with a dataset containing sales data from different regions but it may contain duplicate entries due to various reasons such as multiple source, data entry errors, system glitches. Using pandas funs, how do you remove duplicate entries to ensure data integrity of analysis.

```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East'],
    'Product': ['A', 'B', 'C', 'D', 'A', 'B', 'C'],
    'Sales': [100, 150, 200, 120, 100, 160, 200]
}
df=pd.DataFrame(data)
```

```python
df
```

```python
df.duplicated()
```

drop duplicates for all cols

```python
df2=df.drop_duplicates()
df2
```

```python
duplicate_rows = df[df.duplicated()]
duplicate_rows
```

```python
num_duplicate_rows = df.duplicated().sum()
num_duplicate_rows
```

drop duplicates for select cols only

```python
df3 = df.drop_duplicates(subset=['Region', 'Product'])
df3
```
