write content of excel file 1&2 to excel sheet file 3 sheet 1&2

```python
stocks_data = {
    'Date': ['2022-01-01', '2022-01-02', '2022-01-03'],
    'Symbol': ['AAPL', 'GOOGL', 'MSFT'],
    'Price': [150.25, 2800.0, 350.75],
    'Volume': [1000000, 1500000, 800000],
}
df_stocks = pd.DataFrame(stocks_data)

# Sample data for df_weather
weather_data = {
    'Date': ['2022-01-01', '2022-01-02', '2022-01-03'],
    'Temperature': [25, 28, 30],
    'Humidity': [60, 65, 70],
    'WindSpeed': [10, 12, 15],
}
df_weather = pd.DataFrame(weather_data)
```

```python
with pd.ExcelWriter('/content/drive/MyDrive/datasets/stock2.xlsx') as writer:
  df_stocks.to_excel(writer,sheet_name="stock1.xlsx")
  df_weather.to_excel(writer,sheet_name="weather1.xlsx")
```
